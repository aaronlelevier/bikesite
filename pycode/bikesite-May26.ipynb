```python
import csv

attr = 'td'

def get_content(attr):
    return getattr(row, attr).text.strip() #.replace(' ', '').replace('\n', '')

def trim(attr):
    return getattr(row, attr).text.strip().replace(' ', '').replace('\n', '')

with open(os.path.join(vitalmtb_dir, f'{bike}-3.csv'), 'w', newline='') as f:
    fieldnames = ['Key', 'Value']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    
    for i, row in enumerate(rows):
        if hasattr(row, attr) and hasattr(row.th, 'text'):
            key, value = get_content('th'), trim('td')
            writer.writerow({'Key': key, 'Value': value})
```
